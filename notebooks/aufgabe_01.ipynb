```python
import os

import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt
from IPython.display import Image
from networkx.drawing.nx_agraph import to_agraph
from random import choice

# Ensure the assets directory exists
os.makedirs('assets', exist_ok=True)

prefix = "aufgabe_01"
asset_dir = os.path.join('assets', prefix)
os.makedirs(asset_dir, exist_ok=True)
```

# Aufgabe 1

```python
A = nx.Graph()
A.add_edges_from([(1, 3), (2, 3), (3, 4), (2, 5), (4, 6), (5, 6)])
a = to_agraph(A)

A_filepath = os.path.join(asset_dir, f"A.png")
a.draw(A_filepath, format="png", prog="dot")
display(Image(filename=A_filepath))

a.get_edge(5, 6).attr['color'] = 'red'
for n in [1, 2, 4]:
    a.get_node(n).attr['color'] = 'blue'

for n in [3, 5, 6]:
    a.get_node(n).attr['color'] = 'green'
    
A_colored_filepath = os.path.join(asset_dir, f"A_colored.png")
a.draw(A_colored_filepath, format="png", prog="dot")
display(Image(filename=A_colored_filepath))
```

```python
B = nx.Graph()
B.add_edges_from([(1, 2), (2, 3)])
B.add_node(4)
b = to_agraph(B)
B_filepath = os.path.join(asset_dir, f"B.png")
b.draw(B_filepath, format="png", prog="dot")
display(Image(filename=B_filepath))

for n in [1, 3]:
    b.get_node(n).attr['color'] = 'blue'

for n in [2, 4]:
    b.get_node(n).attr['color'] = 'green'

B_colored_filepath = os.path.join(asset_dir, f"B_colored.png")
b.draw(B_colored_filepath, format="png", prog="dot")
display(Image(filename=B_colored_filepath))
```

```python
C = nx.Graph()
C.add_edges_from([(1, 3), (2, 6), (2, 11), (3, 5), (3, 4), (4, 6), (5, 6), (6, 7), (7, 8), (8, 9), (8, 10), (10, 11)])
c = to_agraph(C)
C_filepath = os.path.join(asset_dir, f"C.png")
c.draw(C_filepath, format="png", prog="dot")
display(Image(filename=C_filepath))

for n in [1, 2, 4, 5, 7, 9, 10]:
    c.get_node(n).attr['color'] = 'blue'

for n in [3, 6, 8, 11]:
    c.get_node(n).attr['color'] = 'green'

C_colored_filepath = os.path.join(asset_dir, f"C_colored.png")
c.draw(C_colored_filepath, format="png", prog="dot")
display(Image(filename=C_colored_filepath))
```
